```python
import numpy as np
class NaiveBayes:
    def __init__(self):
      pass

    def get_stats(self,x):
      mean = np.mean(x, axis=0)
      variance = np.std(x, axis=0, ddof=1)
      print("Mean: " ,mean)
      print("Variance: ",variance)
      return mean,variance

    def pdf(self,x,mean,variance):
      return 1/(variance*np.sqrt(2*np.pi))*np.exp((-1/2)*(x-mean)**2/variance**2)

    def fit(self, X_train, y_train):
      x_0 = X_train[y_train==0]
      x_1 = X_train[y_train==1]
      x_0_stats = self.get_stats(x_0)
      x_1_stats = self.get_stats(x_1)
      self.means = [x_0_stats[0],x_1_stats[0]]
      self.var = [x_0_stats[1], x_1_stats[1]]
      number_of_data_points = X_train.shape[0]
      number_of_x_0 = x_0.shape[0]
      number_of_x_1 = x_1.shape[0]
      x_0_count = len(x_0)
      self.apriori = number_of_x_0 / number_of_data_points, number_of_x_1 / number_of_data_points
      print("apriori:", self.apriori)
      return self.means,self.var,self.apriori

    def predict_proba(self,X):
      self.probs = []
      for i in range(X.shape[0]):
        p_0 = self.apriori[0]
        p_1 = self.apriori[1]
        for j in range(X.shape[1]):
          p_0 = p_0 * self.pdf(X[i][j], self.means[0][j], self.var[0][j])
          p_1 = p_1 * self.pdf(X[i][j], self.means[1][j], self.var[1][j])
        self.probs.append([p_0, p_1])
      return self.probs

    def predict(self,X):
      probs = self.predict_proba(X)
      for i in range(X.shape[0]):
        if probs[i][0] < probs[i][1]:
          print(1)
        else:
          print(0)
```

```python
x = (np.array([2, 3, 1, 4, 0, 2, 3, 2]).reshape(2, 4)).T
y = np.array([0, 1, 1, 0])

print(x)
nb = NaiveBayes()
nb.fit(x, y)
nb.predict(np.array([[0, 3], [10, 4]]))
```
